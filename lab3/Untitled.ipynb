```python
%matplotlib notebook
import sympy
from sympy import Symbol as Symb
from sympy import Function, Piecewise
import random
import numpy as np
import math
from matplotlib import pyplot as plt
import os
import subprocess
PIPE = subprocess.PIPE
```

```python
%env OMP_NUM_THREADS=4
```

```python
def run(n):
    p = subprocess.Popen('/home/daneel/prog/avs/lab3/lab3 {}'.format(n), 
                         shell=True, stdin=PIPE, stdout=PIPE, stderr=subprocess.STDOUT, close_fds=True)
    t1 = float(p.stdout.readline())
    t2 = float(p.stdout.readline())
    return t1, t2
```

```python
times1 = []
times2 = []
for i in range(1, 20000, 100):
    t1, t2 = run(i)
    times1.append(t1)
    times2.append(t2)
```

```python
plt.figure()
plt.xlabel('Size of vector')
plt.ylabel('Time, second')
ls = np.linspace(1, 20000, 200)
plt.plot(ls, times1,  color="blue",label='Direct')
plt.plot(ls, times2,  color="red",label='OMP')
plt.legend()
```
